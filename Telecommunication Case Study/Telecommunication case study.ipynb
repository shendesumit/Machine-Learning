## 2.Identify the Target variable

### Identifying the inputs (x) and output (y)

```python
df.columns
```

```python
# Churn is target/output variable denoted by y
y = df['Churn']

# These columns are input variables denoted by x 
x = df[['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges']]
```

### Target variable is already given so this is supervise learning and the task is Classification Task because given target/output variable is categorical.
### Evaluation metrics is Accuracy.



## 3.split into train and test

```python
# split into train and test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=0)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
print(x_train.shape, y_train.shape)

print(x_test.shape, y_test.shape)
```

### seperating Categorical and Numerical column

```python
x_train.dtypes
```

```python
x_train_cat = x_train.select_dtypes(include=['object'])

x_train_cat.head()
```

```python
x_train_num = x_train.select_dtypes(include=['int64', 'float64'])

x_train_num.head()
```

## 4. Data preparation on train data

### Applying OneHotEncoding to categorical column

```python
x_train_num.head()
```

```python
# scaling the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_num_rescaled = pd.DataFrame(scaler.fit_transform(x_train_num),
                       columns = x_train_num.columns,index = x_train_num.index)

x_train_num_rescaled.head()
```

```python
x_train_cat.head()
```

```python
# OneHotEncoding the categorical features

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first',sparse=False)

x_train_cat_ohe = pd.DataFrame(encoder.fit_transform(x_train_cat), 
                               columns=encoder.get_feature_names(x_train_cat.columns), 
                               index = x_train_cat.index)
x_train_cat_ohe.head()
```

### Concatinating the Encoded Categorical Features and Rescaled Numerical Features

```python
x_train_transformed = pd.concat([x_train_num_rescaled, x_train_cat_ohe], axis=1)

x_train_transformed.head()
```

## 5.preparing Test data

```python
x_test.head()
```

```python
x_test.info()
```

```python
x_test.columns
```

```python
x_test_cat = x_test.select_dtypes(include=['object'])

x_test_cat.head()
```

```python
x_test_cat.index
```

```python
x_test_cat.columns
```

```python
x_test_num = x_test.select_dtypes(include=['int64', 'float64'])

x_test_num.head()
```

```python
x_test_num_rescaled = pd.DataFrame(scaler.transform(x_test_num), 
                                   columns = x_test_num.columns, 
                                   index = x_test_num.index)

x_test_num_rescaled.head()
```

## OHE for Test data

```python
x_test_cat_ohe = pd.DataFrame(encoder.transform(x_test_cat), 
                               columns=encoder.get_feature_names(x_test_cat.columns), 
                               index = x_test_cat.index)
x_test_cat_ohe.head()
```

```python
x_test_transformed = pd.concat([x_test_num_rescaled, x_test_cat_ohe], axis=1)

x_test_transformed.head()
```

## Logistic Regression

```python
from sklearn import metrics
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train_transformed, y_train)
```

```python
y_test_pred =  reg.predict(x_test_transformed)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})
temp_df.head()
```

```python
metrics.accuracy_score(y_test,y_test_pred)
```
